* Let's find out api structure to get data from IDB(https://data.iadb.org/NumbersForDevelopment/NumbersForDevelopment)

* Country :Basic call structure is as shown below.
* 'http://api-data.iadb.org/metadata/country?searchtype=&searchvalue=&languagecode=&responsetype='
* Where, searchtype is code or name:
* Searchvalue: is the value to search (Code/Name/All)
* Responsetype is xml or json
* Languagecode can be en / es

## Querying  metadata. 

* We can find the code and name of all the metadata by Using ''http://api-data.iadb.org/metadata/METADATA?searchtype=code&searchvalue=all&languagecode=en&responsetype=json' as a url by converting METADATA section to country|topic|subtopic|indicator


```python
import requests
import json
import pprint
pp = pprint.PrettyPrinter(indent=4)
metaData = requests.get('http://api-data.iadb.org/metadata/country?searchtype=name&searchvalue=all&languagecode=en&responsetype=json')
allCountries = json.loads(metaData.text)
pp.pprint(allCountries[:10])
```

```python
metaData2 = requests.get( 'http://api-data.iadb.org/metadata/indicator?searchtype=code&searchvalue=all&languagecode=en&responsetype=json')
allIndicator = json.loads(metaData2.text)
pp.pprint(allIndicator[:5])
```

```python
metaData3 = requests.get( 'http://api-data.iadb.org/metadata/topic?searchtype=code&searchvalue=all&languagecode=en&responsetype=json')
allTopics = json.loads(metaData3.text)
pp.pprint(allTopics[:10])
```

```python
metaData4 = requests.get('http://api-data.iadb.org/metadata/subtopic?searchtype=code&searchvalue=all&languagecode=en&responsetype=json')
allSubTopics =json.loads(metaData4.text)
pp.pprint(allSubTopics[:10])
```

## Querying Data

* In the IDB document about data API we can find out we can get data through combining following parameters. 
    *  1)	indicatorcode 
    *  2)	countrycode 
    *  3)	topiccode  
    *  4)	subtopiccode 
    *  5)	frequency 
    *  6)	timeperiod  
    *  7)	Languagecode 
    *  8)	Responsetype 

* We can find out we can get all the data by setting indicatorcode to all

#### To find out the possibility of omitting the data when we set indicatorcode parameter as "all" compare results by few testing. 

```python
#frequencyLevel -> dataString
def queryString(frequencyLevel):
    frequencyDict = {}
    frequencyDict['year'] = "&frequency=year"
    frequencyDict['quarter'] ="&frequency=quarter"
    frequencyDict['month'] = "&frequency=month"
    frequencyDict['all'] = ""
    dataUrl = 'http://api-data.iadb.org/datasitedata?'\
                  'indicatorcode=all&countrycode=ARG'\
                  '{}&languagecode=en&responsetype=json'.format(frequencyDict[frequencyLevel])
    data = requests.get(dataUrl)
    if data.status_code == 200: 
        return json.loads(data.text)
        
        

yearData = queryString("year")
quarterData = queryString("quarter")
monthData = queryString("month")
allData = queryString("all")
assert(len(allData) == len(yearData) + len(quarterData) + len(monthData))
```

```python
def queryByIndicator(indicatorString):
    indicatorCode = "&indicatorCode=" + indicatorString
    dataUrl = 'http://api-data.iadb.org/datasitedata?'\
                  'indicatorcode={}&countrycode=ARG'\
                  '&languagecode=en&responsetype=json'.format(indicatorCode)
    data = requests.get(dataUrl)
    if data.status_code == 200: 
        return json.loads(data.text)
    else: 
        return None
    
metaData2 = requests.get( 'http://api-data.iadb.org/metadata/indicator?searchtype=code&searchvalue=all&languagecode=en&responsetype=json')
allIndicator = json.loads(metaData2.text)
allIndicatorNameList = [ x['IndicatorCode'] for x in allIndicator ]

allIndicatorList = queryByIndicator("all")


indivCounter = 0 
for indicatorName in allIndicatorNameList: 
    indJson = queryByIndicator(indicatorName)
    if indJson:
        indivCounter += len(indivCounter)

assert(len(allIndicatorList) == indivCounter)

```
